```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
print(plt.style.available)
```

```python
plt.style.use(['dark_background'])

ticks = np.arange(3)
liam_scores = [20, 7, 5]
liam_total = 20 + 7 + 5
max_scores = [28, 5, 14]
max_total = 28 + 5 + 14
width = 0.40


plt.bar(ticks, max_scores, width, label='With Picture')
plt.bar(ticks + width, liam_scores, width, label='Control')

plt.xticks(ticks + width / 2, ('Original', 'Prequel', 'Sequel'))
plt.title('Response Frequencies By Group')
plt.ylabel('Frequencies')
plt.legend(loc='best')
plt.show()
```

```python
liam_freq = [liam_scores[0]/liam_total*100, liam_scores[1]/liam_total*100, liam_scores[2]/liam_total*100]
max_freq = [max_scores[0]/max_total*100, max_scores[1]/max_total*100, max_scores[2]/max_total*100]

plt.bar(ticks, max_freq, width, label='With Picture')
plt.bar(ticks + width, liam_freq, width, label='Control')

plt.xticks(ticks + width / 2, ('Original', 'Prequel', 'Sequel'))
plt.title('Percent of Responses By Group')
plt.ylabel('Percent of Responses (%)')
plt.legend(loc='best')
plt.show()
```

```python
print(liam_total, max_total)
```

```python
import random


results = []
#Loop 10000 times, generate random integer (0,3), if equals 3 count = count + 1, repeat for count_2, add count_1-count_2 to results, else continue
for i in range(10000):
    count_1 = 0
    count_2 = 0
    for i in range(max_total+1):
        if random.randint(1, 3) == 3:
            count_1 +=1
    for i in range(max_total+1):
        if random.randint(1, 3) == 3:
            count_2 +=1
    results.append(count_1-count_2)
```

```python
results
```

```python
plt.style.use('seaborn')

plt.hist(results, max_total)
plt.title('Randomly Generated Results')
plt.ylabel('Frequency')
plt.xlabel('Random Expiremental Result - Random Control Result')
plt.show()
```

```python
np_results = np.array(results)
std = np_results.std()
mean = np_results.mean()
(std, mean)
```
